```python
import pandas as pd
import glob

df = glob.glob(r'C:\Users\Aamir Bin Abbas\Desktop\Projects and thesis\Pandas\COVID-19 Global Data\archive\*.csv')

df_list = [pd.read_csv(file) for file in df]
df = pd.concat(df_list, ignore_index=True)

df.head(5)

```

```python
df.info()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df
```

```python
df['province_state'].fillna('Unknown', inplace=True)
df['country_region'].fillna('Unknown', inplace=True)
df['who_region'].fillna('Unknown', inplace=True)
```

```python
df['confirmed'].fillna(0, inplace=True)
df['deaths'].fillna(0, inplace=True)
df['recovered'].fillna(0, inplace=True)
```

```python
df['active'] = df['confirmed'] - df['deaths'] - df['recovered']
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df.groupby('country/region')['confirmed'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('country/region')['deaths'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('country/region')['active'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('date')['confirmed'].sum().plot(title='Global Confirmed Cases Over Time')
```
