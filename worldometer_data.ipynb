```python
import pandas as pd

file_name ='worldometer_data.csv'
cleaned_name = 'worldometer_data_cleaned.csv'

df = pd.read_csv(file_name)

df.head(5)
```

```python
df.set_index('Country/Region',inplace=True)
```

```python
df.sort_values('TotalDeaths',ascending=False,inplace=True)
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
cols = ['Continent', 'Population', 'TotalCases', 'NewCases', 'TotalDeaths',
       'NewDeaths', 'TotalRecovered', 'NewRecovered', 'ActiveCases',
       'Serious,Critical', 'Tot Cases/1M pop', 'Deaths/1M pop', 'TotalTests',
       'Tests/1M pop', 'WHO Region']

df[cols] = df[cols].apply(lambda x : x.fillna(0) if x.dtype in ['int64','float64'] else x.fillna('Unknown'))
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df.to_csv(cleaned_name)
```
